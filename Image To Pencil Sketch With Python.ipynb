```python
import cv2
```

```python
from PIL import Image
from IPython.display import display
```

Reading the image and displaying it

In BGR Format

```python
img=cv2.imread("Flower.png")
img=cv2.resize(img,(625,345))
display(Image.fromarray(img))
```

In RGB Format

```python
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img_rgb))
```

Creating a new image by converting the original image to greyscale

```python
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_img))
```

Inverting the new grayscale image

```python
inverted_img=255-gray_img
display(Image.fromarray(inverted_img))
```

Blurring the image by using the Gaussian Function in OpenCV

```python
blurred = cv2.GaussianBlur(inverted_img, (21, 21), 0)
display(Image.fromarray(blurred))
```

Inverting the blurred image and getting pencil sketch

```python
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_img, inverted_blurred, scale=256.0)
display(Image.fromarray(pencil_sketch))
```

Both original image and pencil sketch

```python
display(Image.fromarray(img_rgb))
display(Image.fromarray(pencil_sketch))
```
